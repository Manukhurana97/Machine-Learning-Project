# NLP -> Sentimental Analysis (Stock depandent on news)

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('Stock_with_sentimen_talanalysis.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
df.describe()
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
len(train), len(test)
```

```python
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
```

```python
data.head()
```

```python
#  Replacing columns name


data.columns = [str(i) for i in range(25)]
data.head()
```

```python
#  convert to small characters

for i in data.columns:
    data[i] = data[i].str.lower()
data.head()
```

```python
all = []
for i in range(len(data.index)):
    all.append(''.join(str(x) for x in data.iloc[i,0:25]))
    
```

```python
all
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#  convert the sentases into vecture of features
vectorcount = CountVectorizer(ngram_range=[2,2])
traindata = vectorcount.fit_transform(all)
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(traindata,train['Label'])
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = vectorcount.transform(test_transform)
predictions = clf.predict(test_dataset)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy = confusion_matrix(test['Label'], predictions)
accuracy
```

```python
score = accuracy_score(test['Label'], predictions)
score
```
