```python
import numpy as np
import pandas  as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
```

```python
df = pd.read_excel('titanic.xls')
df.head(5)
```

```python
orignal_data = df
```

```python
df.drop(['name','body'],1,inplace=True)
df.fillna(0, inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
def convert_to_int(df):
    columns = df.columns.values
   
    for col in columns:
        text_to_list = {}
        
        def get_value(val):
            return text_to_list[val]
        
        if df[col].dtype!=np.int64 and df[col].dtype!=np.float64:
            element = df[col].values.tolist()
            unique_element =set(element)
            
            X=0
            for unique in unique_element:
                if unique not in text_to_list:
                    text_to_list[unique] = X
                    X+=1
            
            df[col] = list(map(get_value, df[col]))
    return df
            

convert_to_int(df)
```

```python
df.fillna(0,inplace=True)
```

```python
X = np.array(df.drop(['survived'],1))
y = np.array(df['survived'])
```

```python
clf = MeanShift()
```

```python
clf.fit(X)
```

```python
labels = clf.labels_
labels
```

```python
cluster_center = clf.cluster_centers_
cluster_center
```

```python
no_of_labels =len(set(labels)) 
no_of_labels
```

```python
orignal_data['cluster_group'] = np.nan
```

```python
for i in range(len(X)):
    orignal_data['cluster_group'].iloc[i]=labels[i]
print(len(orignal_data['cluster_group']))
```

```python
survival_rates = {}

for i in range(no_of_labels):
#   (orignal_data['cluster_group']==float(i))-> True,False; orignal_data[(orignal_data['cluster_group']==float(i))] -> Takes True condition element only
    temp = orignal_data[(orignal_data['cluster_group']==float(i))]
    
#   (temp['survived']==1)-> True ,False; temp[(temp['survived']==1)] Takes True condition element only 
    survival_cluster = temp[(temp['survived']==1)]
    
    survival_rate = len(survival_cluster)/len(temp)
    
    survival_rates[i] =survival_rate
print(survival_rates)
```
