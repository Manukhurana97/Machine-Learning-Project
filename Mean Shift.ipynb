```python
import numpy as np
from sklearn.cluster import MeanShift
import pandas as pd
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
from sklearn.datasets.samples_generator import make_blobs
```

```python
initial_center=[[1,1,1], [3,3,3],[5,5,5]]
```

```python
X,_ = make_blobs(n_samples=1000,centers =initial_center)
```

```python
clf = MeanShift()
clf.fit(X)
```

```python
labels =clf.labels_
labels
```

```python
cluster_center = clf.cluster_centers_
cluster_center
```

```python
n_cluster = len(np.unique(labels))
n_cluster
```

```python
color = 10*['r', 'g', 'c', 'k', 'y', 'm']
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
for i in range(len(X)):
    ax.scatter(X[i][0],X[i][1],X[i][2], c = color[labels[i]])
ax.scatter(cluster_center[:, 0], cluster_center[:, 1], cluster_center[:, 2], marker='X', color='k', zorder=10)
plt.show()
```
